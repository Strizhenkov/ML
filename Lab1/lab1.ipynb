```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from lxml import html
pd.set_option('display.max_rows', None)
url = 'https://en.wikipedia.org/wiki/Opinion_polling_for_the_2024_United_Kingdom_general_election'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
tables = soup.find_all('table', class_='wikitable sortable collapsible mw-datatable')
allData = []
headers = ['Year']
for i in tables[0].find_all('th'):
    headers.append(i.text[:-1])    #skip \n at the string's endings.
headers = headers[:-7]    #skip 6 whitespaces in list at the end.

#parsing tables

for table in tables[:4]:
    for i in table.find_all('tr')[4:]:    #skip wrong strings
        row_data = i.find_all('td')
        row = [2024 - tables.index(table)] + [j.text[:-1] for j in row_data[:-1]]
        lastColumnValue = row[-1][:row[-1].find('%') + 1]    #delete useless information from last column
        row = row[:-1]
        if (len(row) == 12) and (row[5] != '—') and (row[6] != 'TBC') and (row[7] != 'TBC') and (row[8] != 'TBC') and (row[9] != 'TBC') and (row[10] != 'TBC') and (row[11] != 'TBC') and (lastColumnValue != 'TBC'):    #skip wrong rows
            row.append(lastColumnValue)
            row[5] = row[5][:row[5].find(',')] + row[5][row[5].find(',') + 1:]
            for j in range(6, 13):     #convet % into float values 
                if (row[j].find('%') == -1):    #— == 0% == 0
                    row[j] = '0'
                else:
                    row[j] = str(float(row[j][:row[j].find('%')]) / 100)    #ex. 35% == 0.35
            allData.append(tuple(row))

#create data frame and convert some colums

myData = pd.DataFrame(allData, columns = headers)
myData = myData.astype({'Sample size' : int})
myData = myData.astype({'Con.' : float, 'Con.' : float, 'Lab.' : float, 'Lib. Dems' : float, 'SNP' : float, 'Green' : float, 'Ref.' : float, 'Others' : float})
myData.loc[:]
```
